```python
from sklearn.datasets import load_iris
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import KMeans
```

```python
data = load_iris()
```

```python
list(data)
```

```python
a = np.array(data['data'])
a[:,2]
```

```python
distx = squareform(pdist(data['data'], metric='euclidean'))

R = dendrogram(linkage(distx, method='single'), color_threshold=3)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
kmeans = KMeans(4)
#randomstate is set to match results
Y_hat_kmeans = kmeans.fit(data['data']).labels_

plt.scatter(a[:,1], a[:,2],c=Y_hat_kmeans);
```
