```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading predownloaded CSV file, low_memory=False was set cause of multiple type patterns
verisdata = pd.read_csv('data/vcdb.csv',low_memory=False)
```

```python
#see if the data loaded
verisdata.head()
```

```python
#playing around with column display
verisdata[['victim.victim_id','victim.state']].head()
```

```python
#setting incident_id as index
verisdata = verisdata.set_index('incident_id')
```

```python
#printing to make sure index setting worked
verisdata.head()
```

```python
by_country = verisdata.groupby('victim.country')
```

```python
by_country.size().order(ascending=True)
```

```python
#plotting out number of incidnets per country
by_country.size().order(ascending=False).plot(logy=True,figsize=(8, 6))
```

```python
#looking at distributino of how incidents are discovered
verisdata.groupby('discovery_method').size().order(ascending=False).plot(kind='bar',figsize=(20, 6))
```

```python
#looking at distribution of how incidents are discovered
discoverymethod = verisdata.groupby('discovery_method').size().order(ascending=False)
```

```python
discoverymethod = pd.DataFrame(discoverymethod)
```
